```python
from sklearn.datasets import load_diabetes
import matplotlib.pyplot as plt
diabetes = load_diabetes()
print(diabetes.data.shape, diabetes.target.shape)
```

```python
diabetes.data[0:3]
```

```python
plt.scatter(diabetes.data[:,2],diabetes.target)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = diabetes.data[:,2]
y = diabetes.target
print(x)
print(y[0])
```

```python
w = 1.0
b = 1.0
```

```python
y = x[0] * w + b
print(y)
```

```python
print(y[0])
```

```python
# 변화율로 가중치 업데이트하기

w_hat = w + 0.1
y_hat = x[0] * w_hat + b
print(y_hat)
w_rate = (y_hat-y)/(w_hat-w)
print(w_rate)
```

```python
# 변화율로 절편 업데이트하기

b_hat = b + 0.1
y_hat = x[0] * w + b_hat
print(y_hat)
b_rate = (y_hat-y)/(b_hat-b)
print(b_rate)
```

```python
# 오차 역전파로 가중치와 절편 업데이트

w = 1.0
b = 1.0
y_hat = x[0] * w + b
err = y[0] - y_hat
w_new = w + w_rate *err
b_new = b + 1*err
print(w_new,b_new)
```

```python
for x_i,y_i in zip(x,y):
    y_hat = x_i *w + b
    err = y_i - y_hat
    w_rate = x_i
    w = w + w_rate *err
    b = b + 1*err
print(w,b)
```
